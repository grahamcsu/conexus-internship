# Q1
[Question 1](https://projecteuler.net/problem=1)

```python
x=0
sum=0
iter=0
while iter<1000:
    if x%3==0 or x%5==0:
        sum=sum+x
    iter=iter+1
    x=x+1
print(sum)
```

# Q3
[Question 3](https://projecteuler.net/problem=3)

```python
import math
x=1
n=600851475143
factors=[]
factors.append(1)
if n%2==0:
    factors.append(2)
x=3
while x*x<=n:
    while n%x==0:
        factors.append(x)
        n//=x
    x=x+2
print(factors)
```

# Q4
[Question 4](https://projecteuler.net/problem=4)

```python
num1= 999
num2=999
def is_palindrome(x):
    rev_x=x[::-1]
    if x==rev_x:
        return True
    else:
        return False
while is_palindrome(str(num1*num2))==False:
    num1=num1-1
    num2=999
    while num2>900:
        if is_palindrome(str(num2*num1)):
            break
        else:
            num2=num2-1
print(num1*num2)
```
