# Overview
This notebook provides a basic introduction to scientific computing using `numpy` and visualization with `seaborn` and `matplotlib`, focusing on trigonometric functions.

# Imports

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Basic Sinusoid Anatomy

```python
# Define the x range
x = np.linspace(0, 2*np.pi, 1000)

# Define the wave parameters (change these)
A = 2            # amplitude
omega = 1        # angular frequency
phi = np.pi / 4  # phase shift
C = 0            # vertical shift

# Define the sinusoid
y = A * np.sin(omega * x + phi) + C
```

```python
sns.set_theme(style="darkgrid", context="notebook")

plt.figure(figsize=(10, 5))
ax = sns.lineplot(x=x, y=y, color="crimson", linewidth=2.5)

ax.set_title(f"Sinusoid: $y = A\\,\\sin(\\omega x + \\phi)+C$", fontsize=15, pad=12)
ax.set_xlabel("x (radians)", fontsize=12)
ax.set_ylabel("y", fontsize=12)
ax.axhline(0, color="gray", linewidth=0.8, linestyle="--")

sns.despine()
plt.tight_layout()
plt.show()
```

# Adding Sinusoids

```python
# Define the two waves (edit these)
y1 = 1 * np.sin(1 * x + 0)
y2 = 1 * np.sin(1 * x + np.pi)

# Plot the waves and their sum
sns.set_theme(style="darkgrid", context="notebook")
plt.figure(figsize=(11, 5))
ax = sns.lineplot(x=x, y=y1, color="steelblue", linewidth=1.5, label="wave 1")
sns.lineplot(x=x, y=y2, color="seagreen", linewidth=1.5, label="wave 2", ax=ax)
sns.lineplot(x=x, y=y1 + y2, color="crimson", linewidth=2.5, label="sum", ax=ax)
ax.axhline(0, color="gray", linewidth=0.8, linestyle="--")
ax.set(title="Wave 1 + Wave 2", xlabel="x (radians)", ylabel="y")
sns.despine()
plt.show()
```

# The Unit Circle and Sine

Sine and cosine are the coordinates of a point traveling around a circle of radius 1. As the angle $\theta$ sweeps from $0$ to $2\pi$, the point's height traces out the sine wave.

```python
theta = np.pi / 3  # pick an angle (try changing this)

sns.set_theme(style="darkgrid", context="notebook")
fig, (left, right) = plt.subplots(1, 2, figsize=(13, 5))

# Left: the unit circle with the point at angle theta
circle_t = np.linspace(0, 2*np.pi, 1000)
left.plot(np.cos(circle_t), np.sin(circle_t), color="gray")      # the circle itself
left.plot([0, np.cos(theta)], [0, np.sin(theta)], color="crimson")  # radius to the point
left.scatter(np.cos(theta), np.sin(theta), color="crimson", zorder=3)
left.set_aspect("equal")  # keep the circle round, not oval
left.set(title=f"Unit circle at θ = {theta:.2f}", xlabel="cos θ", ylabel="sin θ")

# Right: sine wave, with the same point marked
right.plot(x, np.sin(x), color="steelblue")
right.scatter(theta, np.sin(theta), color="crimson", zorder=3)
right.axhline(np.sin(theta), color="crimson", linewidth=0.8, linestyle="--")
right.set(title="sin θ", xlabel="θ (radians)", ylabel="height")

sns.despine()
plt.tight_layout()
plt.show()
```

# Numerically validate the Pythagorean identity

The Pythagorean identity says $\sin^2\theta + \cos^2\theta = 1$ for every angle. Instead of proving it, we can verify it numerically across thousands of points.

```python
# Compute both sides at every x at once
identity = np.sin(x)**2 + np.cos(x)**2

# np.allclose checks "equal to 1 everywhere" while tolerating tiny rounding error
print("Always equal to 1?", np.allclose(identity, 1))

# Plot it: a flat line at y = 1 confirms the identity holds for all x
sns.set_theme(style="darkgrid", context="notebook")
plt.figure(figsize=(11, 4))
sns.lineplot(x=x, y=identity, color="crimson", linewidth=2)
plt.ylim(0, 2)  
plt.title(r"$\sin^2 x + \cos^2 x$")
plt.xlabel("x (radians)"); plt.ylabel("value")
sns.despine()
plt.show()
```
